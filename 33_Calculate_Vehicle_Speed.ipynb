```python
import cv2 
import numpy as np
import time

# Define the region of interest coordinates
ax1, ax2 = 10, 550
bx1, bx2 = 200, 450
ay, by = 400, 250
i = 1

def calcSpeed(start_time, end_time):
    diff = end_time - start_time
    try:
        speed = (  5.001 / 1000) / (diff / 3600) 
        return speed
    except ZeroDivisionError:
        print("Can't divide by zero")
        return 0

# Load the cascade classifier
car_cascade_classifier = r'haarcascades/cars.xml'
car_cascade = cv2.CascadeClassifier(car_cascade_classifier)

# Load the video
cap = cv2.VideoCapture('videos/highwayCamera.mp4')

# Main loop
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    roi_gray = gray[:, 300:850]
    roi = frame[:, 300:850]

    bfilter = cv2.bilateralFilter(roi_gray, 11, 17, 17)
    cars = car_cascade.detectMultiScale(bfilter, 1.1, 2)

    # Draw the lines
    cv2.line(roi, (ax1, ay), (ax2, ay), (176, 30, 150), 2)
    cv2.line(roi, (bx1, by), (bx2, by), (176, 30, 150), 2)

    for (x, y, w, h) in cars:
        # Draw the rectangle around detected cars
        cv2.rectangle(roi, (x, y), (x + w, y + h), (0, 255, 0), 2)
        center_y = int((y + y + h) / 2)

        # Check if the car crosses the first line (ay)
        if int(ay) == center_y:
            start_time = time.time()
            cv2.line(roi, (ax1, ay), (ax2, ay), (176, 20, 0), 2)
            print('Car detected at line ay')

        # Check if the car crosses the second line (by)
        if int(by) <= center_y <= int(by + 10):
            end_time = time.time()
            speed = calcSpeed(start_time, end_time)
            print(f'Car number {i}, speed = {speed:.2f} km/h')
            i += 1
            cv2.line(roi, (bx1, by), (bx2, by), (176, 20, 0), 2)
        else :
                cv2.putText(roi, "Calcuting", (100,200), cv2.FONT_HERSHEY_SIMPLEX,                       

1,(0,255,0),3)                            

                break     

    cv2.imshow('ROI', roi)
    if cv2.waitKey(60) & 0xFF == 27:  # Press ESC to exit
        break

cap.release()
cv2.destroyAllWindows()
```
