```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('images/sudoku_game.webp',0)
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

laplcian = cv2.Laplacian(img , cv2.CV_64F , ksize=3)
laplcian = np.uint8(np.absolute(laplcian))
sobelx = cv2.Sobel(img , cv2.CV_64F , 1 ,0)
sobely = cv2.Sobel(img , cv2.CV_64F , 0 ,1)
combinedSobel = cv2.bitwise_or(sobelx,sobely)

title = ['Original image' , 'Laplacian' ,'sobelx' ,'sobely' ,'Combined Sobel']
images = [img, laplcian ,sobelx ,sobely ,combinedSobel]

plt.figure(figsize=(10,8))
for i in range(len(title)):
    plt.subplot(2,3, i + 1)
    plt.imshow(images[i])
    plt.title(title[i])
    plt.xticks([])
    plt.yticks([])
```
