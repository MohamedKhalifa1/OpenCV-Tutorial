```python
import cv2 
import numpy as np
from time import sleep


ww = 80
hh = 80
offest = 6
y1 = 550
delay = 60
carros = 0
detec = []


def page_center(x,y,w,h):
    x1 = int(w//2)
    y1 = int(h//2)

    cx = x1 + x
    cy = y1 + y

    return cx , cy


cap = cv2.VideoCapture('videos/highwayCamera.mp4')


subtractor =  cv2.createBackgroundSubtractorMOG2(history=100, varThreshold=50)

while cap.isOpened():
    ret , frame1 = cap.read()
    gray = cv2.cvtColor(frame1 , cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray , (3,3) , 5)
    img_sub = subtractor.apply(blur)
    dialte = cv2.dilate(img_sub , np.ones((3,3)))


    contours , _ = cv2.findContours(dialte , cv2.RETR_TREE , cv2.CHAIN_APPROX_NONE)
    cv2.line(frame1 , (30 , y1) , (1100,y1) , (176 , 30 ,150) , 3)

    for (i , c) in enumerate(contours):
        (x,y,w,h) = cv2.boundingRect(c)

        valid_contour = (w >= ww) and (h >= hh)

        if not valid_contour :
            continue

        cv2.rectangle(frame1 , (x ,y) , (x + w , y + h) , (0 ,255 ,0) , 2)

        center = page_center(x,y,w,h)
        detec.append(center)
        cv2.circle(frame1 , center , 4 , (0,0,255) , -1)

        for (x , y) in detec:
            if (y < (y1 + offest) and y > (y1 - offest) ):
                carros += 1
                cv2.line(frame1 , (30 , y1) , (1100,y1) , (0 , 127 ,255) , 3)
                detec.remove((x,y))
                print('No. of cars detected :'+ str(carros))
    cv2.putText(frame1 , 'No. of cars detected :' + str(carros) , (370 , 70) , cv2.FONT_HERSHEY_SCRIPT_SIMPLEX , 2 , (0,0,255) , 4)

    cv2.imshow("Video Original", frame1)
    cv2.imshow(" Detectar ", dialte)

    if cv2.waitKey(1) == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
