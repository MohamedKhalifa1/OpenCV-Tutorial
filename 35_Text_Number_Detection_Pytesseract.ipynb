```python
import pytesseract
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('images/wedding card.jpg')
img_gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

plt.imshow(cv2.cvtColor(img_gray , cv2.COLOR_BGR2RGB))
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
boxes = pytesseract.image_to_boxes(img_gray)
height, width, _ = img.shape  

for box in boxes.splitlines():
    box = box.split()
    char, x, y, w, h = box[0], int(box[1]), int(box[2]), int(box[3]), int(box[4])
    
    
    y = height - y
    h = height - h
    
    cv2.rectangle(img, (x, h), (w, y), (0, 0, 255), 2)
    cv2.putText(img , char , ((x, y+25)) ,cv2.FONT_HERSHEY_PLAIN , 1,(0,26,25) , 2 )

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
config = r'--oem 3 --psm 6 outputbase digits'
boxes = pytesseract.image_to_boxes(img_gray , config=config)
height, width, _ = img.shape  

for box in boxes.splitlines():
    box = box.split()
    char, x, y, w, h = box[0], int(box[1]), int(box[2]), int(box[3]), int(box[4])
    
    
    y = height - y
    h = height - h
    
    cv2.rectangle(img, (x, h), (w, y), (0, 0, 255), 2)
    cv2.putText(img , char , ((x, y+25)) ,cv2.FONT_HERSHEY_PLAIN , 1,(0,26,25) , 2 )

cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
config = r'--oem 3 --psm 6 outputbase digits'
boxes = pytesseract.image_to_data(img_gray )
height, width, _ = img.shape  

for i , b in enumerate(boxes.splitlines()):
    if i!=0:
        b = b.split()
        if len(b)==12:
            x,y,w , h = int(b[6]) ,int(b[7])  , int(b[8]) ,int(b[9])
           
            text = b[11]
            cv2.rectangle(img, (x, y), (x+ w, y+h), (0, 0, 255), 2)
            cv2.putText(img , text , ((x, y+25)) ,cv2.FONT_HERSHEY_PLAIN , 1,(0,26,25) , 2 )
cv2.imshow('imge', img)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```
